```python
import pandas as pd
from mordred import Calculator, descriptors
from rdkit import Chem
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("data/qmof_property_cell.csv", index_col=0)
```

```python
data.columns
```

```python
calc = Calculator(descriptors)
```

```python
mol = Chem.MolFromSmiles("[O-]C(=O)c1cc(C)c(cc1C)C(=O)[O-]")
d = calc.pandas([mol, mol])
```

```python
mordred_descriptors = pd.DataFrame(columns=d.columns)
for linkers, filename in zip(data.smilesLinkers, data.filename):
    try:
        descriptors_for_linkers = calc.pandas([Chem.MolFromSmiles(linker) for linker in linkers.split(",")])
        mordred_descriptors.loc[filename] = descriptors_for_linkers.mean()
    except TypeError:
        print(filename)
```

```python
mordred_descriptors.to_csv("data/mordred_descriptors.csv")
```
